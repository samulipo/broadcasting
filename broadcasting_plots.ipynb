```python
### ADDING NOISE TO A LANDSCAPE. WITH 20-amplitude noise ###

bumpVals = np.arange(start=0,stop=301,step=10)

## EWORK PLOT
#IMPORT DATA
eWork400IND=pd.read_csv('adding_bumps/ind_eWorkAt1000_addingBumps.csv', sep=',',header=None)
eWork400SOC=pd.read_csv('adding_bumps/soc_eWorkAt1000_addingBumps.csv', sep=',',header=None)
eWork400UNI=pd.read_csv('adding_bumps/uni_eWorkAt1000_addingBumps.csv', sep=',',header=None)

plt.figure()

plt.plot(bumpVals,eWork400IND,color='black',linestyle='-', label='individual learners')
plt.plot(bumpVals,eWork400SOC,color='black',linestyle='--',label='social learners')
plt.plot(bumpVals,eWork400UNI,color='black',linestyle=':',label='diverse population')

plt.legend(loc='upper right')
plt.xlabel(r'Ruggedness ($\beta$)')
plt.ylabel('Proportion of epistemic work done at t=1000')
#plt.suptitle('Noise added.')
plt.xlim((0,300))
plt.ylim((0,0.7))
plt.savefig('adding_bumps/apr2016_eWork_1000_adding_noise.pdf')

plt.show()

## EPROG PLOT
#IMPORT DATA
eProg400IND=pd.read_csv('adding_bumps/ind_eProgAt1000_addingBumps.csv', sep=',',header=None)
eProg400SOC=pd.read_csv('adding_bumps/soc_eProgAt1000_addingBumps.csv', sep=',',header=None)
eProg400UNI=pd.read_csv('adding_bumps/uni_eProgAt1000_addingBumps.csv', sep=',',header=None)

plt.figure()

plt.plot(bumpVals,eProg400IND,color='black',linestyle='-', label='individual learners')
plt.plot(bumpVals,eProg400SOC,color='black',linestyle='--', label='social learners')
plt.plot(bumpVals,eProg400UNI,color='black',linestyle=':', label='diverse population')

plt.legend(loc='upper right')
plt.xlabel(r'Ruggedness ($\beta$)')
plt.ylabel('Epistemic progress (*) at t=1000')
plt.xlim((0,300))
plt.ylim((0,1.0))
plt.savefig('adding_bumps/apr2016_eProg_1000_adding_noise.pdf')
```
